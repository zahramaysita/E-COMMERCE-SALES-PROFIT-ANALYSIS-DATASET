### **Outliers Values**

Outliers adalah nilai yang jauh berbeda dari nilai lainnya dalam dataset. Nilai Outlier bisa jauh lebih rendah atau lebih tinggi. Outlier bisa terjadi karena berbagai alasan seperti faktor kesalahan maupun kejadian lain yang tidak terduga.

Pada tahap ini, kita akan melihat nilai ekstrem (outliers) pada kolom yang bertipe data numerik. Hal ini dilakukan agar mendapatkan insight yang akurat sehingga dapat memperkuat analisis yang akan dilakukan.

Akan tetapi, tidak semua kolom perlu ditangani. Kita tetap harus melihat dan memahami konteks dari setiap kolom yang ada pada dataset.

Dan terlihat kolom:

* Quantity

Dengan persentase 0%

* Sales

Dengan persentase 1%

* Profit

Dengan persentase 5.26%

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

1. **Quantity**

Memiliki persentase outlier sebesar 0%

Karena tidak ditemukan outlier pada kolom Quantity (0%), maka tidak diperlukan penanganan khusus. Hal ini menunjukkan bahwa jumlah barang yang dibeli oleh pelanggan masih berada dalam rentang yang wajar dan tidak terdapat pembelian dalam jumlah yang ekstrem.

Dengan demikian, data pada kolom Quantity tetap digunakan sepenuhnya dalam analisis karena sudah merepresentasikan kondisi transaksi yang normal tanpa adanya nilai yang menyimpang.

2. **Sales**

Memiliki persentase outlier sebesar 1%

Berdasarkan hasil deteksi menggunakan metode IQR (Interquartile Range), ditemukan bahwa terdapat outlier pada kolom Sales sebesar 1% dari total data. Meskipun jumlahnya relatif kecil, nilai tersebut tetap termasuk dalam kategori outlier karena berada di luar batas bawah atau batas atas yang telah ditentukan.

Sesuai dengan teori penanganan outlier yang digunakan, jika jumlah outlier kurang dari atau sama dengan 10% dari total data, maka data tersebut dapat dihapus. Oleh karena itu, dilakukan penghapusan terhadap data outlier pada kolom Sales.

Penghapusan ini bertujuan untuk mengurangi pengaruh nilai penjualan yang terlalu ekstrem terhadap perhitungan statistik seperti rata-rata (mean), sehingga hasil analisis menjadi lebih stabil dan lebih merepresentasikan mayoritas transaksi yang terjadi.

```python
Q1 = df['Sales'].quantile(0.25)
Q3 = df['Sales'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df['Sales'] = df['Sales'].clip(lower=lower_bound, upper=upper_bound)
```

3. **Profit**

Berdasarkan hasil pengecekan, terdapat outlier pada kolom Profit sebesar 5,25% dari total data. Walaupun jumlahnya masih di bawah 10%, nilai profit yang terlalu tinggi atau terlalu rendah bisa memengaruhi perhitungan rata-rata dan membuat hasil analisis menjadi kurang stabil.

Outlier pada Profit bisa terjadi karena adanya diskon besar, promosi, atau perbedaan keuntungan antar produk. Namun, jika dibiarkan begitu saja, nilai yang terlalu ekstrem bisa membuat rata-rata keuntungan terlihat tidak sesuai dengan kondisi mayoritas transaksi.

Oleh karena itu, dilakukan penanganan menggunakan metode IQR (Interquartile Range). Metode ini bekerja dengan menentukan batas bawah dan batas atas berdasarkan sebaran data, sehingga kita bisa membedakan mana nilai yang masih wajar dan mana yang sudah terlalu ekstrem.

```python
Q1 = df['Profit'].quantile(0.25)
Q3 = df['Profit'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df['Profit'] = df['Profit'].clip(lower=lower_bound, upper=upper_bound)
```

4. **Mengecek Kembali Nilai Ekstrem**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Setelah dilakukan penanganan outlier menggunakan metode IQR, boxplot menunjukkan bahwa tidak terdapat lagi nilai penjualan yang berada di luar batas bawah maupun batas atas. Artinya, nilai transaksi yang sebelumnya terlalu ekstrem telah dihapus sehingga distribusi data menjadi lebih stabil.

Sebagian besar nilai Sales kini berada dalam rentang yang lebih terkendali, dan sebaran data terlihat lebih seimbang. Jarak antara kuartil pertama (Q1) dan kuartil ketiga (Q3) menjadi lebih representatif terhadap mayoritas transaksi.

Setelah dilakukan penanganan menggunakan metode IQR, boxplot menunjukkan bahwa tidak lagi terdapat nilai yang berada di luar batas bawah maupun batas atas. Hal ini menandakan bahwa outlier pada kolom Profit telah berhasil ditangani.

Sebelumnya terdapat 5,25% data yang termasuk outlier. Setelah proses penanganan dilakukan, sebaran data menjadi lebih stabil dan tidak ada lagi nilai ekstrem yang dapat memengaruhi perhitungan rata-rata.

Median berada di tengah distribusi dan rentang antar kuartil (IQR) terlihat lebih terkontrol, sehingga data Profit kini lebih representatif terhadap mayoritas transaksi.

# **Construct Data**

1. **Membuat Kolom Baru "Profit Status"**

Kolom Profit_Status dibuat untuk mempermudah melihat apakah suatu transaksi menghasilkan keuntungan atau kerugian. Nilai profit yang awalnya berupa angka diubah menjadi kategori seperti Untung, Impas, dan Rugi agar lebih mudah dibaca dan dipahami.

Dengan adanya kolom ini, kita tidak perlu melihat angka profit satu per satu untuk mengetahui kondisi transaksi. Kita bisa langsung melihat berapa banyak transaksi yang untung dan berapa yang rugi.

```python
def categorize_profit(profit):
    if profit > 0:
        return 'Untung'
    elif profit < 0:
        return 'Rugi'
    else:
        return 'Break Even'

df['Profit Status'] = df['Profit'].apply(categorize_profit)
```

Dengan adanya Profit_Margin, kita tidak hanya melihat besar kecilnya profit dalam bentuk angka, tetapi juga bisa mengetahui apakah suatu transaksi memiliki keuntungan yang tinggi atau rendah secara proporsional.

```python
sales_by_category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
sales_by_category_top3 = sales_by_category.head(3)

plt.figure(figsize=(10, 8))
sales_by_category_top3.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proporsi 3 Kategori Produk Teratas Berdasarkan Total Penjualan')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

## **Distribution/Distribusi**

Aktivitas : Menganalisis distribusi nilai transaksi pada kolom Sales.

Tujuan : Untuk mengetahui pola penyebaran nilai transaksi, melihat apakah mayoritas transaksi bernilai kecil atau besar, serta mengidentifikasi adanya kecenderungan distribusi (skewness) pada data penjualan.

Visualisasi : Histogram.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales'], bins=20, kde=True)
plt.title('Distribusi Penjualan (Sales)')
plt.xlabel('Sales')
plt.ylabel('Frekuensi')
plt.show()
```

Distribusi Sales tidak berbentuk normal karena tidak simetris. Grafik menunjukkan pola right-skewed (positively skewed), dimana sebagian besar transaksi berada pada nilai rendah dan hanya sedikit transaksi dengan nilai sangat tinggi.

## **Relationship/Hubungan**

Aktivitas : Menganalisis hubungan antara nilai penjualan (Sales) dengan keuntungan (Profit).

Tujuan : Untuk mengetahui apakah peningkatan nilai penjualan berbanding lurus dengan peningkatan profit, serta mengidentifikasi kemungkinan adanya transaksi dengan penjualan tinggi namun menghasilkan profit rendah atau bahkan kerugian.

Visualisasi : Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['Sales', 'Profit']].corr(),
            annot=True,
            cmap='coolwarm',
            fmt='.2f')
plt.title('Korelasi antara Sales dan Profit')
plt.show()
```

Heatmap digunakan untuk melihat hubungan antar variabel dalam bentuk korelasi. Pada visualisasi ini, warna merah menunjukkan hubungan positif, sedangkan warna biru menunjukkan hubungan negatif.

Berdasarkan heatmap tersebut, nilai korelasi antara Sales dan Profit adalah 0,83.

Nilai 0,83 menunjukkan bahwa Sales dan Profit memiliki hubungan positif yang kuat. Artinya, ketika penjualan meningkat, keuntungan juga cenderung meningkat.

Warna merah pada bagian diagonal menunjukkan nilai 1,00 karena variabel dibandingkan dengan dirinya sendiri. Sementara itu, nilai 0,83 antara Sales dan Profit menunjukkan hubungan yang kuat dan searah.
